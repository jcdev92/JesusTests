```python
df_1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3'], 
 'C':['C0','C1','C2','C3'], 
 'D':['D0','D1','D2','D3']})

df_2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
 'B':['B4','B5','B6','B7'], 
 'C':['C4','C5','C6','C7'], 
 'D':['D4','D5','D6','D7']})
```

### **MERGE** - COLUMNS MATCH

```python
izq = pd.DataFrame({'Key':['K0','K1','K2','K3'], 
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'Key':['K0','K1','K2','K3'], 
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})
```

```python
izq.merge(der)   # Anteriormente era izq.merge(der, on='Key')
```

```python
izqui = pd.DataFrame({'key':['K0','K1','K2','K3'], 
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']})

dere = pd.DataFrame({'key_2':['K0','K1','K2','K3'], 
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})
```

```python
izqui.merge(dere, left_on='key', right_on='key_2')
```

```python
izqui = pd.DataFrame({'key':['K0','K1','K2','K3'], 
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']})

dere = pd.DataFrame({'key_2':['K0','K1','K2',np.NaN], 
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})
```

```python
dere
```

```python
izqui.merge(dere, left_on='key', right_on='key_2') # didnt get the third row in the merge, becouse in key_2 - k3 was a NaN Value and the rows cant be matched
```

```python
izqui.merge(dere, left_on='key', right_on='key_2', how='right') # forcing get third right row, ignoring left third row.
```

### **JOIN** - INDEX MATCH

```python
iz = pd.DataFrame({'A':['A0','A1','A2'],
 'B':['B0','B1','B2']},
 index=['k0','k1','k2'])
de = pd.DataFrame({'C':['C0','C1','C2'], 
 'D':['D0','D1','D2']}, 
 index=['k0','k2','k3'])
```

```python
iz.join(de)
```

```python
iz.join(de, how='inner')
```

```python
iz.join(de, how='left')
```

```python
iz.join(de, how='right')
```

```python
iz.join(de, how='outer')
```
